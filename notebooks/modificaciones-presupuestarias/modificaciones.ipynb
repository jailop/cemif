```python
import pandas as pd
import sqlite3
```

```python
stmt_aproved = """
    SELECT year, a.office, office_name,
        SUM(approved) AS approved,
        SUM(modified) AS modified,
        SUM(accrued)  AS accrued,
        SUM(shifted) AS shifted
    FROM accrued AS a
    LEFT JOIN office AS b
        ON a.office = b.office
    WHERE a.office LIKE "%00"
    GROUP BY year, a.office
    ORDER BY year, shifted
"""
```

```python
db = sqlite3.connect("../../data/batch2/data/master.db")
```

```python
data = pd.read_sql(stmt_aproved, db)
data["ratio"] = data.shifted / data.approved * 100
data.to_csv("modificaciones.csv", index=False)
```
